```python
#Librerias a utilizar
import camelot
import pandas as pd
```

```python
#PDF a utilizar
file_path = 'solo_tablas.pdf'

# Leer todas las tablas del PDF
tables = camelot.read_pdf(file_path, pages='all', flavor='stream')  # 'stream' funciona bien cuando las tablas no tienen bordes claros
print(f"Se encontraron {len(tables)} tablas")

# Dataframe para guardar todas las tablas
all_data = pd.DataFrame()

# Extraer las primeras 11 columnas de cada tabla
#algunas universidades presentan sus tablas en formatos con variaciones en sus columnas, donde las primeras 11 columnas son las mismas para todas
#las universidades
for table in tables:
    df = table.df
    # Tomamos solo las primeras 11 columnas
    df_11cols = df.iloc[:, :12]
    all_data = pd.concat([all_data, df_11cols], ignore_index=True)
```

```python
#asignamos nombres de columnas
all_data.rename(columns={0:'Codigo',
                         1:'Carrera',
                         2:'Sede',
                         3:'NEM',
                         4:'Ranking',
                         5:'CL',
                         6:'CM1',
                         7:'HyCsS',
                         8:'o',
                         9:'Cs',
                         10:'CM2',
                         11:'Otra'}, inplace=True)
```
